## Phylogenetic Trees

### Background 
Phylogenetic trees provide quantitative and visual representations of the evolutionary relationships between sequences and species. Various algorithms are available for constructing phylogenetic trees. The algorithms differ greatly in their approaches, but the final results are often directly comparable. Today, we will use an online phylogenetic workflow to build and analyze three trees.
